```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def cargar_datos():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
    nombres_columnas = ['longitud_sepalo', 'ancho_sepalo', 'longitud_petalo', 'ancho_petalo', 'clase']
    datos = pd.read_csv(url, names=nombres_columnas)
    return datos

def analizar_datos_basicos(df):
    # Información básica del dataframe
    print(df.info())

    # Descripción estadística de las columnas numéricas
    print(df.describe())

    # Verificar si hay valores nulos
    print("Valores nulos por columna:")
    print(df.isnull().sum())

def visualizar_datos(df):
    # Histogramas
    df.hist(figsize=(10, 10))
    plt.suptitle("Histogramas de las características")
    plt.show()

    # Box plots por característica para cada clase de iris
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x='clase', y='longitud_sepalo')
    plt.title('Boxplot de Longitud de Sétalo por Clase de Iris')
    plt.show()

    # Pair plot para visualizar las relaciones entre características
    sns.pairplot(df, hue='clase')
    plt.show()

def main():
    datos = cargar_datos()
    analizar_datos_basicos(datos)
    visualizar_datos(datos)

if __name__ == "__main__":
    main()
```
